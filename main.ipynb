```python
import os
import pandas as pd
import numpy as np
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python
import os
os.getcwd()
```

```python
def run_ml(df_for_ml, C, penalty, solver):
    X = df_for_ml.iloc[:, 1:-1]
    y = df_for_ml.iloc[:,-1]

    clf = LogisticRegression(random_state=0).fit(X, y)

    #print("For the week of ", last_week)
    acc_score = accuracy_score(clf.predict(X), y)
    bal_acc_score = balanced_accuracy_score(clf.predict(X), y)

    cm = confusion_matrix(y, clf.predict(X))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
    #disp.plot()
    #plt.show()
    
    return acc_score, bal_acc_score, disp
```

### For classes from week 2 and M127

```python
data_folder = "C:\\Users\\hongj\\Data\\Data_incubator\\capstone\\week2"
```

```python
df_for_ml_list = []

x_axis = [*range(2, 12, 1)]

for i in x_axis:
    frames = [pd.read_csv(data_folder + '\M127_week' + str(i) + '_fall2019.csv'), 
            pd.read_csv(data_folder + '\M127_week' + str(i) + '_fall2020.csv')]
    
    df_for_ml = pd.concat(frames)
    df_for_ml = df_for_ml.reset_index()
    del df_for_ml['index']

    df_for_ml_list.append(df_for_ml)
```

```python
df_for_ml_list[0]
```

```python
acc_score_list = []
bal_score_list = []

for i, item in zip(x_axis, df_for_ml_list):
    acc_score, bal_acc_score, disp = run_ml(item, C = 1, penalty = 'none', solver = 'newton-cg')
    print("For the week of", i)
    print("Accuracy Score is", acc_score)
    print("Balanced Accuracy Score is", bal_acc_score)
    
    acc_score_list.append(acc_score)
    bal_score_list.append(bal_acc_score)
    
    disp.plot()
    
    disp.figure_.savefig('lower_week2_confusion_matrix_' + str(i) + '.png')
    plt.show()
    print("-" * 75)
```

```python
plt.plot(x_axis, acc_score_list, label = 'acc score')
plt.plot(x_axis, bal_score_list, label = 'bal acc score')
plt.xlabel('Weeks')
plt.legend(loc="lower right")
plt.savefig('plot_week2_low.png')
plt.show()
```

### For the classes from week 3 and M181, M126, M127

```python
data_folder = "C:\\Users\\hongj\\Data\\Data_incubator\\capstone\\week3"
```

```python
df_for_ml_list = []

x_axis = [*range(3, 12, 1)]

for i in x_axis:
    frames = [pd.read_csv(data_folder + '\M127_week' + str(i) + '_fall2021.csv'),
            pd.read_csv(data_folder + '\M127_week' + str(i) + '_spr2022.csv'),
            pd.read_csv(data_folder + '\M126_week' + str(i) + '.csv'),
            pd.read_csv(data_folder + '\M181_week' + str(i) + '.csv') ]
    
    df_for_ml = pd.concat(frames)
    df_for_ml = df_for_ml.reset_index()
    del df_for_ml['index']

    df_for_ml_list.append(df_for_ml)
```

```python
df_for_ml_list[0]
```

```python
df_for_ml_list[1]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
```

```python
for df in df_for_ml_list:
    df_for_ml = df

    X = df_for_ml.iloc[:, 1:-1]
    y = df_for_ml.iloc[:,-1]

    logit = LogisticRegression()
    linear_model = Pipeline([
        ('logistic', logit)
    ])

    param_grid = {
        "logistic__C" : [0.01, 0.1, 0.5, 1, 2], 
        "logistic__penalty" : ['l1', 'l2', 'elasticnet', 'none'],
        "logistic__solver" :['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
    }

    search = GridSearchCV(linear_model, param_grid, n_jobs=2, cv = 5)
    search.fit(X, y)
    print("Best parameter (CV score=%0.3f):" % search.best_score_)
    print(search.best_params_)
```

```python
acc_score_list = []
bal_score_list = []

for i, item in zip(x_axis, df_for_ml_list):
    acc_score, bal_acc_score, disp = run_ml(item, C = 0.01, penalty = 'none', solver = 'newton-cg')
    print("For the week of", i)
    print("Accuracy Score is", acc_score)
    print("Balanced Accuracy Score is", bal_acc_score)
    
    acc_score_list.append(acc_score)
    bal_score_list.append(bal_acc_score)
    
    disp.plot()
    disp.figure_.savefig('lower_week3_confusion_matrix_' + str(i) + '.png')
    plt.show()
    print("-" * 75)
```

```python
print(acc_score_list)
print(bal_score_list)
```

```python
plt.plot(x_axis, acc_score_list, label = 'acc score')
plt.plot(x_axis, bal_score_list, label = 'bal acc score')
plt.xlabel('Weeks')
plt.legend(loc="lower right")
plt.savefig('plot_week3_low.png')
plt.show()
```
